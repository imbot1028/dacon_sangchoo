```python

# train input 데이터 전처리(case_1 ~ case_28)
nums = range(1,29)
train_input_murge = pd.DataFrame()
for num in nums:
    if num < 10 :
        df = pd.read_csv('./train_input/CASE_0%d.csv'%num)
        df['Case'] = 'Case%d'%num
        train_input_murge = pd.concat([df, train_input_murge])
        
    else :
        df = pd.read_csv('./train_input/CASE_%d.csv'%num)
        df['Case'] = 'Case%d'%num
        train_input_murge = pd.concat([df, train_input_murge])

        
train_input = train_input_murge.sort_values(['Case','DAT']).copy()
```

```python

#학습셋 target merge
train_target_merge = pd.DataFrame()
nums = range(1,29)
for num in nums:
    if num < 10 :
        df = pd.read_csv('./train_target/CASE_0%d.csv'%num)
        df['Case'] = 'Case%d'%num
        train_target_merge = pd.concat([df, train_target_merge])
    else :
        df = pd.read_csv('./train_target/CASE_%d.csv'%num)
        df['Case'] = 'Case%d'%num
        train_target_merge = pd.concat([train_target_merge, df])
        
train_target = train_target_merge.copy()
train_target.reset_index(drop=True)
```

```python
train_case02 = pd.read_csv('./train_input/CASE_02.csv')
train_case02[['시간당백색광량','시간당청색광량','시간당적색광량','시간당총광량']] = train_case02[['시간당백색광량','시간당청색광량','시간당적색광량','시간당총광량']].replace(0,np.NaN)
# 데이터 전처리 코드 시간배열을 일단위로 변경
# full data는 평균치로 처리
exvar_data = train_case02.groupby('DAT').agg(temp = ('내부온도관측치','mean'), humidity = ('내부습도관측치', 'mean'), co2 = ('내부습도관측치', 'mean'), ec = ('ec관측치', 'mean'))
# 광량데이터는 평균 및 일별 광량 시간으로 처리
led_sum_data = train_case02.groupby('DAT').agg(white_led_sum = ('시간당백색광량', 'sum'), blue_led_sum = ('시간당청색광량', 'sum'), red_led_sum = ('시간당적색광량', 'sum'), all_led_sum = ('시간당총광량', 'sum'))
# led_count 합량
led_count_data = train_case02.groupby('DAT').agg(white_led_count = ('시간당백색광량', 'count'), blue_led_count = ('시간당청색광량', 'count'), red_led_count = ('시간당적색광량', 'count'), all_led_count = ('시간당총광량', 'count'))
train_case02_target = pd.read_csv('./train_target/CASE_02.csv').drop(columns = 'DAT')
```

```python
ml = pd.concat([exvar_data, led_sum_data, led_count_data], axis=1)
ml
```
